```python
def stan_currency(curr):
    if curr == '$$' or curr == 'Dollars':
        return 'USD'
    return curr

df_financials =pd.read_excel('movies_db.xlsx', 'financials', converters= {'currency': stan_currency})
```

```python
df_financials.head()
```

```python
df_merged = pd.merge(df_movie,df_financials, on="movie_id")
df_merged.head()
```

```python
def grouper(df, idx, col):
    if 80 <= df[col].loc[idx] <= 90:
        return '80-90'
    if 50 <= df[col].loc[idx] <= 60:
        return '50-60'
```

```python
gg = df.groupby(lambda idx : grouper(df, idx, 'temperature'))
```

```python
gg
```

```python
for key, d in gg:
    print("key :" , key)
    print("______")
    print("data :", d)
```
